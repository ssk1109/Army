# list []
+ 순서를 가진 객체의 집합
+ 모든 자료형 다 넣을 수 있음

````python
apple = ["a", "p", "p", "l", "e"]
print(apple)
print("")

# 뒤에서부터 추가
apple.append("s")
print(apple)
print("")

#원하는 위치에 추가
apple.insert(2,"wow")
print(apple)
print("")

# 뒤에서 부터 빼기
print(f"빠진 문자: {apple.pop()}")
print(apple)
print("")

# 같은 값 찾기
print(apple.count("p"))
print("")

# 정렬
apple.sort()
print(apple)
# 바로 정렬
nums = [3,5,2,12,6,100]
print(sorted(nums)) # sorted를 사용한다고 원본이 바로 정렬되는 것은 아님. 그저 정렬된 값만 보여줄 뿐.
print(nums)
print("")

# 뒤집기
apple.reverse()
print(apple)
# 이걸 이터레이터 객체라고 함.
print(reversed(nums))
print(list(reversed(nums))) # reversed는 data를 바로 보여주는 것이 아니기 때문에 list()로 감싸주어야함.
# 마찬가지로 원본은 반전 안됨~
print(nums)
print("")

# sort(키, 역순여부)를 결정할 수 있음
nums.sort(reverse=True) # 역순
print(nums)
# key= abs(절댓값), len(문자열 시 길이) 등
nums.sort(key=abs)
print(nums)
print("")

# 리스트 합치기
apple.extend(nums)
print(apple)
print("")

# 모두 삭제
apple.clear()
print(apple)

````

# 함수
```
def 함수명(매개변수):
    실행문
    return 반환값 # 반환값은 1개 이상 가능, 반환값은 없어도 되긴함.
```

````python
def deposit(balance, money):
  balance+=money
  print("입금완료! 잔액: {}원".format(balance))
  return balance

def withDraw(balance, money):
  if balance>=money:
    print("출금완료! 잔액: {}원".format(balance-money))
    return balance - money
  else:
    print("출금불가! 잔액: {}원".format(balance))
  return balance

def withDraw_night(balance, money):
  commission = 100
  if balance>=money+commission:
    print("출금완료! 수수료: {}원, 잔액: {}원".format(commission, balance-money-commission))
    return commission, balance - money - commission
  else:
    print("출금불가! 잔액: {}원".format(balance))
  return balance


balance = 0
balance = deposit(balance, 1000)
balance = withDraw(balance, 500)
balance = withDraw(balance, 1000)
commission, balance = withDraw_night(balance, 200) # 반환값이 2개 이상이기 때문에 순서대로 받을 수 있음
print(commission, balance)
````

#지역변수, 전역변수
+ 함수 내에서 전역변수를 사용하기 위해 변수 안에 전역변수라는 표시 필요.
```
global 변수명
```
+ but 함수 전달값으로 계산하고 반환값을 사용하는 방식이 정배.



````python
num =10

def adder(n):
  global num # 전역변수!
  num = num + n
  print(f"함수 안 {num}")

adder(5)
print(f"함수 밖 {num}")
````

````python
num = 10

def adder(num, n):
  num +=n
  print(f"함수 안 {num}")
  return num

num = adder(num, 10)
print(f"함수 밖 {num}")
````

# 출력포멧
+ 정수형
  + {0:빈자리부호 | 정렬방향 | 부호여부 | 자릿수 | 세자리끊기}
+ 유리수
  + {0:.반올림할자릿수f}

````python
# 빈 자리는 빈공간으로 두고, 오른쪽 정렬을 하되, 총 10자리 공간 확보
print("{0: >10}".format(500))

# 양수일 때는 +, 음수일 때는 -
print("{0: >+10}".format(500))
print("{0: >+10}".format(-500))

# 왼쪽 정렬을 하고, 빈칸을 _로 채움
print("{0:_<+10}".format(500))

# 세 자리마다 , 찍어주기
print("{0:,}".format(100000000000))

# 세 자리마다 , 찍어주되 +-까지 표현하기
print("{0:+,}".format(100000000000))
print("{0:+,}".format(-100000000000))

# 세 자리마다 , 찍어주되 부호표시, 자릿수 확보, 빈자리는 ^
print("{0:^<+30,}".format(100000000000))

# 소숫점 출력
print("{0:f}".format(5/3))

# 소숫점 자릿수 표시
print("{0:.2f}".format(5/3))
````

````python
class Unit:
# __init__
  def __init__(self, name, hp, damage): # class안의 메소드 앞에는 self를 항상 적어놔야함.
    self.name =name
    self.hp = hp
    self.damage = damage
    print("{} 유닛이 생성 되었습니다.".format(self.name))
    print("체력: {}, 공격력: {}".format(self.hp,self.damage))

# 메소드
class Attack_unit:
  def __init__(self, name, hp, damage):
    self.name =name # self.name 중 name은 멤버변수
    self.hp = hp # self.hp 중 hp은 멤버변수
    self.damage = damage # self.damage 중 damage은 멤버변수

  def attack(self, location):
    print("{} : {} 방향으로 적군을 공격합니다. [공격력 {}]".format(self.name,location, self.damage))
    # self.name, self.damage는 __init__에서 정의된 name, damage를 사용.
    # location은 attack 함수의 location 사용.

  def damaged(self, damage):
    print("{} : {} 데미지를 입었습니다.".format(self.name, damage))
    self.hp -= damage
    if(self.hp<=0):
      print("{} : 파괴되었습니다".format(self.name))
    else:
      print("{} 남은 체력 : {}".format(self.name, self.hp))

#class
marine1 = Unit("marine1", 40, 5)
marine2 = Unit("marine2", 40, 5)
tank = Unit("tank", 150, 35)
print()

# 멤버변수
wraith1 = Unit("wraith", 80, 5)
# .을 사용하여 class 외부에서 class의 멤버변수 사용 가능
print("이름: {}, 공격력: {}".format(wraith1.name, wraith1.damage))

#멤버변수
wraith2 = Unit("taken wraith", 80, 5)
#clocking이라는 변수는 class에 없는 변수를 추가로 할당 가능.
wraith2.clocking=True
if(wraith2.clocking==True):
  # 아무리 wraith2에 clocking이라는 변수가 할당되었다고는 해도 wraith1까지에게는 미치지 않음.
  # -> 확장을 한 겍체에 대해서만 적용
  print("{}은 이미 클로킹 상태입니다!".format(wraith2.name))
print()

# 메소드
firebat1 = Attack_unit("firebat", 50, 16)
firebat1.attack("1시")
firebat1.damaged(25)
firebat1.damaged(25)
````

# 다중상속
+ 다수의 부모 class

````python
class Unit:
  def __init__(self, name, hp):
    self.name = name
    self.hp = hp

class Attack_unit:
  def __init__(self, name, hp, damage):
    Unit.__init__(self, name, hp)
    self.damage = damage

  def attack(self, location):
    print("{} : {} 방향으로 적군을 공격합니다. [공격력 {}]".format(self.name,location, self.damage))

  def damaged(self, damage):
    print("{} : {} 데미지를 입었습니다.".format(self.name, damage))
    self.hp -= damage
    print(f"{self.name} : 남은 체력 : {self.hp}")
    if(self.hp<=0):
      print("{} : 파괴되었습니다".format(self.name))

class Air_unit:
  def __init__(self, air_speed):
    self.air_speed = air_speed

  def fly(self, name, location):
    print(f"{name} : {location} 방향으로 비행합니다. [비행속도 {self.air_speed}]")

class Air_Attack_unit(Attack_unit, Air_unit): # 두 class를 상속
  def __init__(self, name, hp, damage, air_speed):
    Attack_unit.__init__(self, name, hp, damage)
    Air_unit.__init__(self, air_speed)

valkyrie = Air_Attack_unit("발키리", 200, 6, 5)
valkyrie.fly(valkyrie.name, "서쪽")
````

# 메소드 오버라이팅
+ 부모 class에서 정의한 method 말고 자식 class에서 정의한 method를 사용할 때.
+ 부모 class의 method 그대로 사용하는 것이 아닌 자식 class에서 새로 정의한 것을 쓰겠다!
+ method 덮어쓰기

````python
class Unit:
    def __init__(self, name, hp, speed):
      self.name = name
      self.hp = hp
      self.speed = speed

    def move(self, location):
      print("[지상유닛]")
      print("{} : {} 방향으로 이동 [속도 {}]".format(self.name, location, self.speed))

class Attack_unit(Unit):
    def __init__(self, name, hp, speed, damage):
      Unit.__init__(self, name, hp, speed)
      self.damage = damage

    def attack(self, location):
      print("{} : {} 방향으로 적군을 공격합니다. [공격력 {}]".format(self.name,location, self.damage))

    def damaged(self, damage):
      print("{} : {} 데미지를 입었습니다.".format(self.name, damage))
      self.hp -= damage
      print(f"{self.name} : 남은 체력 : {self.hp}")
      if(self.hp<=0):
        print("{} : 파괴되었습니다".format(self.name))

class Air_unit:
    def __init__(self, air_speed):
      self.air_speed = air_speed

    def fly(self, name, location):
      print(f"{name} : {location} 방향으로 비행합니다. [비행속도 {self.air_speed}]")

class Air_Attack_unit(Attack_unit, Air_unit):
    def __init__(self, name, hp, damage, air_speed):
      Attack_unit.__init__(self, name, hp, 0, damage) # 공중유닛이기 때문에 공중 속도는 0으로 처리.
      Air_unit.__init__(self, air_speed)

    # Unit class의 move()함수를 오버라이딩 한 상황!
    def move(self, location):
      print("[공중유닛]")
      self.fly(self.name, location)


valture = Attack_unit("벌처", 80, 10, 20)

battle_curiser = Air_Attack_unit("배틀크루저", 500, 25, 3)

valture.move("북서쪽")
battle_curiser.move("남동쪽")
````

````python
class Unit:
    def __init__(self, name, hp, speed):
      self.name = name
      self.hp = hp
      self.speed = speed

class Building_unit(Unit):
  def __init__(self, name, hp, location):
    # 아래 방식으로 Unit의 __init__을 사용해서 초기화를 해도 되지만
    # Unit.__init__(self, name, hp, 0)
    # super를 사용하여 초기화를 해도 된다!
    super().__init__(name, hp, 0) # super를 쓸 때는 super 뒤에 괄호를 붙이고 self를 적지 않는다
    self.location = location
````

````python
from random import *

# 일반 유닛
class Unit:
    def __init__(self, name, hp, speed):
        self.name = name
        self.hp = hp
        self.speed = speed
        print("{0} 유닛이 생성되었습니다.".format(name))

    def move(self, location):
        print("{0} : {1} 방향으로 이동합니다. [속도 {2}]"\
            .format(self.name, location, self.speed))

    def damaged(self, damage):
        print("{0} : {1} 데미지를 입었습니다.".format(self.name, damage))
        self.hp -= damage
        print("{0} : 현재 체력은 {1} 입니다.".format(self.name, self.hp))
        if self.hp <= 0:
            print("{0} : 파괴되었습니다.".format(self.name))

# 공격 유닛
class AttackUnit(Unit):
    def __init__(self, name, hp, speed, damage):
        Unit.__init__(self, name, hp, speed)
        self.damage = damage

    def attack(self, location):
        print("{0} : {1} 방향으로 적군을 공격 합니다. [공격력 {2}]" \
            .format(self.name, location, self.damage))

# 마린
class Marine(AttackUnit):
    def __init__(self):
        AttackUnit.__init__(self, "마린", 40, 1, 5) # 이름, 체력, 이동속도, 공격력

    # 스팀팩 : 일정 시간 동안 이동 및 공격 속도를 증가, 체력 10 감소
    def stimpack(self):
        if self.hp > 10:
            self.hp -= 10
            print("{0} : 스팀팩을 사용합니다. (HP 10 감소)".format(self.name))
        else:
            print("{0} : 체력이 부족하여 스팀팩을 사용하지 않습니다".format(self.name))

# 탱크
class Tank(AttackUnit):
    # 시즈모드 : 탱크를 지상에 고정시켜, 더 높은 파워로 공격 가능. 이동 불가.
    siege_developed = False # 시즈모드 개발여부 (클래스 변수)

    def __init__(self):
        AttackUnit.__init__(self, "탱크", 150, 1, 35) # 이름, 체력, 이동속도, 공격력
        self.siege_mode = False # 시즈모드 (해제 상태)

    # 시즈모드
    def set_siege_mode(self):
        if Tank.siege_developed == False: # 시즈모드가 개발되지 않은 경우 메소드 탈출
            return

        # 현재 시즈모드가 아닐 때
        if self.siege_mode == False:
            print("{0} : 시즈모드로 전환합니다.".format(self.name))
            self.damage *= 2 # 공격력 2배로 증가
            self.siege_mode = True # 시즈 모드 설정
        # 현재 시즈모드일 때
        else:
            print("{0} : 시즈모드를 해제합니다.".format(self.name))
            self.damage /= 2 # 공격력 절반으로 감소
            self.siege_mode = False # 시즈 모드 해제

# 날 수 있는 기능을 가진 클래스
class Flyable:
    def __init__(self, flying_speed):
        self.flying_speed = flying_speed

    def fly(self, name, location):
        print("{0} : {1} 방향으로 날아갑니다. [속도 {2}]"\
            .format(name, location, self.flying_speed))

# 공중 공격 유닛
class FlyableAttackUnit(AttackUnit, Flyable):
    def __init__(self, name, hp, damage, flying_speed):
        AttackUnit.__init__(self, name, hp, 0, damage)
        Flyable.__init__(self, flying_speed)

    def move(self, location):
        self.fly(self.name, location)

# 레이스
class Wraith(FlyableAttackUnit):
    def __init__(self):
        FlyableAttackUnit.__init__(self, "레이스", 80, 20, 5) # 체력, 공격력, 공중 이동 속도
        self.cloaked = False # 클로킹 모드 (해제 상태)

    # 클로킹 모드
    def cloaking(self):
        # 현재 클로킹 모드일 때
        if self.cloaked == True:
            print("{0} : 클로킹 모드 해제합니다.".format(self.name))
            self.cloaked = False
        # 현재 클로킹 모드가 아닐 때
        else:
            print("{0} : 클로킹 모드 설정합니다.".format(self.name))
            self.cloaked = True

# 게임 시작
def game_start():
    print("[알림] 새로운 게임을 시작합니다.")

# 게임 종료
def game_over():
    print("Player : gg") # good game
    print("[Player] 님이 게임에서 퇴장하셨습니다.")

# 실제 게임 진행
game_start()

# 마린 3기 생성
m1 = Marine()
m2 = Marine()
m3 = Marine()

# 탱크 2기 생성
t1 = Tank()
t2 = Tank()

# 레이스 1기 생성
w1 = Wraith()

# 유닛 일괄 관리 (생성된 모든 유닛 append)
attack_units = []
attack_units.append(m1)
attack_units.append(m2)
attack_units.append(m3)
attack_units.append(t1)
attack_units.append(t1)
attack_units.append(w1)

# 전군 이동
for unit in attack_units:
    unit.move("1시")

# 탱크 시즈모드 개발
Tank.siege_developed = True
print("[알림] 탱크 시즈 모드 개발이 완료되었습니다.")

# 공격 모드 준비 (마린 : 스팀팩, 탱크 : 시즈모드, 레이스 : 클로킹)
for unit in attack_units:
    if isinstance(unit, Marine): # Marine 의 인스턴스이면 스팀팩
        unit.stimpack()
    elif isinstance(unit, Tank): # Tank 의 인스턴스이면 시즈모드
        unit.set_siege_mode()
    elif isinstance(unit, Wraith): # Wraith 의 인스턴스이면 클로킹
        unit.cloaking()

# 전군 공격
for unit in attack_units:
    unit.attack("1시")

# 전군 피해
for unit in attack_units:
    unit.damaged(randint(5, 20)) # 공격은 랜덤으로 받음 (5 ~ 20)

# 게임 종료
game_over()
````

# 예외처리
+ try 부분을 실행하다 오류가 발생할 시 except 부분의 특정 오류에 대한 부분을 실행
```
try:
  실행문
except err1:
  err1에 대한 예외처리
except err2 as err: # 발생한 err 내용 확인
  err2에 대한 예외처리
  print(err)
except Exception as err:
  정의하지 않은 모든 err에 대한 예외처리
  print(err) # 이런 식으로 err 내용도 확인 가능
```

````python
try:
  print("나누기 전용 계산기")
  a, b = map(int, input().split())

  print(f"{a} / {b} = {int(a/b)}")
except ValueError:
  print("Error!")
except ZeroDivisionError as err:
  print(err)
except Exception as err:
  print("Unknown Error!")
  print(err)
````

# 사용자 정의 예외처리
+ 사용자 정의 에러 만들기
```
class err1(Exception): # err1 == 만들 에러명
  pass

try:
  실행문
  raise err1
except err1:
  예외처리
```
+ 사용자 정의 에러에 대한 msg 출력
```
class err2(exception)
  def __init__(self, msg):
    self.msg = msg
  def __str__(self)
    return self.msg

try:
  실행문
  raise err2("에러 발생시 나올 msg")
except err2 as err:
  예외처리
  print(err) # msg 출력
```

````python
class BigNumberError(Exception):
  def __init__(self, msg):
    self.msg = msg

  def __str__(self):
    return self.msg

try:
  print("한 자리 수 나누기 전용 계산기")
  a, b = map(int, input().split())
  if(a>=10 or b>=10):
    raise BigNumberError(f"입력값: {a}, {b}")
  print(f"{a} / {b} = {int(a/b)}")
except ValueError:
  print("한 자리 수를 입력하세요!")
except BigNumberError as err:
  print("한 자리 숫자만 입력하세요")
  print(err)
````

+ 오류 발생 시 예외처리를 통해 err가 났을 때 프로그램이 강제 종료 되는 것을 막는다!
+ error 종류: 없는 파일 열기, list에 없는 값 접근 등

````python
class BigNumberError(Exception):
  def __init__(self, msg):
    self.msg = msg

  def __str__(self):
    return self.msg

try:
  print("한 자리 수 나누기 전용 계산기")
  a, b = map(int, input().split())
  if(a>=10 or b>=10):
    raise BigNumberError(f"입력값: {a}, {b}")
  print(f"{a} / {b} = {int(a/b)}")
except ValueError:
  print("한 자리 수를 입력하세요!")
except BigNumberError as err:
  print("한 자리 숫자만 입력하세요")
  print(err)
finally:
  print("계산기 이용 종료!")
````

````python
from re import S
class SoldOutError(Exception):
  def __init__(self, msg):
    self.msg = msg
  def __str__(self):
    return self.msg

chicken = 10
waiting = 1

while(True):
  try:
    print(f"[남은 치킨 : {chicken}]")
    order = int(input("치킨 몇 마리 주문? "))
    if(order<=0):
      raise ValueError
    elif(order > chicken):
      print("재료부족")
    else:
      print(f"[대기번호 {waiting}] {order}마리 주문이 완료되었습니다.")
      waiting +=1
      chicken -= order
    if(chicken==0):
      raise SoldOutError("재고가 소진되어 더 이상 주문을 받지 않습니다.")
  except ValueError:
    print("잘못된 값을 입력하였습니다.")
  except SoldOutError as err:
    print(err)
    break
````
