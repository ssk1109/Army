# list []
+ 순서를 가진 객체의 집합
+ 모든 자료형 다 넣을 수 있음

```python
apple = ["a", "p", "p", "l", "e"]
print(apple)
print("")

# 뒤에서부터 추가
apple.append("s")
print(apple)
print("")

#원하는 위치에 추가
apple.insert(2,"wow")
print(apple)
print("")

# 뒤에서 부터 빼기
print(f"빠진 문자: {apple.pop()}")
print(apple)
print("")

# 같은 값 찾기
print(apple.count("p"))
print("")

# 정렬
apple.sort()
print(apple)
# 바로 정렬
nums = [3,5,2,12,6,100]
print(sorted(nums)) # sorted를 사용한다고 원본이 바로 정렬되는 것은 아님. 그저 정렬된 값만 보여줄 뿐.
print(nums)
print("")

# 뒤집기
apple.reverse()
print(apple)
# 이걸 이터레이터 객체라고 함.
print(reversed(nums))
print(list(reversed(nums))) # reversed는 data를 바로 보여주는 것이 아니기 때문에 list()로 감싸주어야함.
# 마찬가지로 원본은 반전 안됨~
print(nums)
print("")

# sort(키, 역순여부)를 결정할 수 있음
nums.sort(reverse=True) # 역순
print(nums)
# key= abs(절댓값), len(문자열 시 길이) 등
nums.sort(key=abs)
print(nums)
print("")

# 리스트 합치기
apple.extend(nums)
print(apple)
print("")

# 모두 삭제
apple.clear()
print(apple)

```

# dictionary {}
+ key-value 관계
+ key값은 str도 가능


```python
name = {1:"신", 2: "성", 3:"규"}
print(name)

# value 하나만 출력
print(name[1])
print(name.get(1))
# []는 없는 key 값을 찾을 때 프로그램 종료
# but .get은 없으면 none이나 원하는 내용 출력 가능
# print(name[4])
print(name.get(4, "이름이 네 글자인 사람도 있나요?"))
print("")

# key 값 찾기
print(1 in name)
print(4 in name)
print("")

# 값 추가
print(name)
name[1] = "박"
name[4] = "네 글자 이름"
print(name)
print("")

# 값 삭제
del name[4]
print(name)
print("")

# key만 출력
print(name.keys())
print("")

# value만 출력
print(name.values())
print("")

# key-value 값으로 출력
print(name.items())
print("")

# dictionary 삭제
name.clear()
print(name)
```
