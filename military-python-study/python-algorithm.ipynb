```python
from collections import Counter

a,b,c=map(int,input().split())
nums = set([a,b,c])
counts=Counter([a,b,c])

if(len(nums)==1):
  print(10000+a*1000)
elif(len(nums)==2):
  if(counts[a]==2):
    print(1000+a*100)
  elif(counts[b]==2):
    print(1000+b*100)
  elif(counts[c]==2):
    print(1000+c*100)
else:
  print(max(a,b,c)*100)
```

```python
a, b, c = map(int, input().split())
if a == b == c:
    print(10000 + a*1000)
elif a == b or a == c:
    print(1000 + a*100)
elif b == c:
    print(1000 + b*100)
else:
    print(max(a, b, c)*100)
```
